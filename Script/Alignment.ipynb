```python
import pandas as pd
import os
```

# Best hit SCCmec extraction

```python
best_hit_scc = []
base_dir = "/Users/marker/Desktop/MRSA_handover/Alignments_for_genome_annotation"
   
cols = [
        "qseqid","qlen","sseqid","slen","pident","length", #Columns for blast results
        "mismatch","gapopen","qstart","qend",
        "sstart","send","evalue","bitscore"
]

for files in os.listdir(base_dir):
    if not files.endswith("_genomic_cassette.tsv"):
        continue
    filepath = os.path.join(base_dir, files)
    absolute_path,extension = os.path.splitext(files) #remove extension from file absolute path
    relative_path = os.path.basename(absolute_path) #get the name without extension
    sample_id = relative_path.split('_')[2] #get sample ID 
    
    df = pd.read_csv(filepath,sep = '\t', header = None, names = cols)
    df['length'] =  df['length'].astype(float)
    df['slen'] =  df['slen'].astype(float)
    df['coverage'] = df['length'] / df['slen'] #calculate coverage
    df_sorted = df.sort_values(['evalue', 'length', 'coverage', 'pident'], ascending=[True, False, False, False]) #prioritize, evalue, alignment legnth, coverage and the %identity
    row = df_sorted.iloc[0] #Extract best hit row
    best_hit_scc.append({
        "Sample_ID": sample_id,
        "SCCmec type": row["sseqid"],
        "SCCmec start": f"{float(row['qstart'])}_{row['qseqid']}", #concatenate contig location
        "SCCmec end": f"{float(row['qend'])}_{row['qseqid']}",
        "SCCmec Reference length": float(row["slen"]),
        "SCCmec Alignment length": float(row["length"]),
        "SCCmec Percent identity": float(row["pident"]),
        "Coverage": float(row['coverage']),
        "SCCmec start val": float(row['qstart']),
        "SCCmec end val": float(row['qend'])
    })

final_df = pd.DataFrame(best_hit_scc) 

    
    
    
    
```

# Best hit single genes extraction

```python
best_hit_singlegenes = []
base_dir = "/Users/marker/Desktop/MRSA_handover/Alignments_for_genome_annotation"
   
cols = [
        "qseqid","qlen","sseqid","slen","pident","length",
        "mismatch","gapopen","qstart","qend",
        "sstart","send","evalue","bitscore"
]

for files in os.listdir(base_dir):
    if not files.endswith("_single-genes.tsv"):
        continue
    filepath = os.path.join(base_dir, files)
    absolute_path,extension = os.path.splitext(files) #remove extension from file absolute path
    relative_path = os.path.basename(absolute_path) #get the name without extension
    sample_id = relative_path.split('_')[2] #get sample ID 
    
    df_single = pd.read_csv(filepath,sep = '\t', header = None, names = cols)
    df_single['gene'] = df_single['sseqid'].str.split(':').str[0] #extract gene name
    df_single['gene_final'] = df_single['gene'].str.split('[-_]').str[0] #special case for ccrC1 gene where does seperate by - or , instead of :
    gene_list = list(df_single['gene_final'].unique()) 
    for i in gene_list: #loop through each gene
        df_copy = df_single.copy()
        df_copy_gene = df_copy[df_copy['gene_final'] == i]
        df_copy_gene['length'] =  df_copy_gene['length'].astype(float)
        df_copy_gene['slen'] =  df_copy_gene['slen'].astype(float)
        df_copy_gene['coverage'] = df_copy_gene['length'] / df_copy_gene['slen'] #calculate coverage
        df_copy_gene_sorted = df_copy_gene.sort_values(['evalue', 'length', 'coverage', 'pident'], ascending=[True, False, False, False]) 
        row = df_copy_gene_sorted.iloc[0]
        best_hit_singlegenes.append({
            "Sample_ID": sample_id,
            "Gene": row["sseqid"],
            "Start": f"{float(row['qstart'])}_{row['qseqid']}",
            "End": f"{float(row['qend'])}_{row['qseqid']}",
            "Reference length": float(row["slen"]),
            "Alignment length": float(row["length"]),
            "Percent identity": float(row["pident"]),
            "Coverage": float(row['coverage']),
            "Gene abv": row['gene_final'],
            "Start val": float(row['qstart']),
            "End val": float(row['qend'])
        })
    
final_df_single_gene = pd.DataFrame(best_hit_singlegenes)
```

```python
orfX_df = final_df_single_gene[final_df_single_gene['Gene abv'] == 'orfX']
orfX_df.rename(columns={'Gene':'OrfX gene','Start':'OrfX start','End':'OrfX end','Reference length':'OrfX reference length',  #renaming before merging
                        'Alignment length':'OrfX Alignment','Percent identity':'OrfX percent identity','Coverage':'OrfX coverage',
                        "Start val":"OrfX start val","End val":"OrfX end val"}, inplace = True)
```

```python
merged_sccme_orfX = pd.merge(final_df,orfX_df,on='Sample_ID',how = 'inner') #merge with SCCmec df
merged_sccme_orfX = merged_sccme_orfX.sort_values('Sample_ID')
```

```python
#Calulate OrfX-SCCmec junction distance
orfx_scc_junc = []
for index,row in merged_sccme_orfX.iterrows():
    if row['OrfX end val'] < row['SCCmec start val']: #foward strand 
        junction_foward = row['SCCmec start val'] - row['OrfX end val'] 
        orfx_scc_junc.append(junction_foward)
    elif row['OrfX start val'] > row['SCCmec end val']:
        junction_reverse= row['OrfX start val'] - row['SCCmec end val'] #Reverse strand
        orfx_scc_junc.append(junction_reverse)
    else: 
        orfx_scc_junc.append('OrfX inside SCCMec')
merged_sccme_orfX['OrfX-SCCmec junction'] = orfx_scc_junc
```
